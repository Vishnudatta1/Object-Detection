```python
import json
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval


# Extract annotations from Model 1
annotations_1 = rgb_coco_format

# Extract annotations from Model 2
annotations_2 = grd_coco_format

# Assign weights to each model's predictions
weight_1 = 2  # Weight for Model 1's predictions
weight_2 = 2  # Weight for Model 2's predictions

# Create a new list for fused predictions
fused_annotations = []

# Combine predictions from Model 1 and Model 2 using weighted score fusion
for pred_1 in annotations_1:
    matching_preds = [pred_2 for pred_2 in annotations_2 if pred_2['image_id'] == pred_1['image_id']]

    if matching_preds:
        pred_2 = matching_preds[0]
        fused_score = weight_1 * pred_1['score'] + weight_2 * pred_2['score']
        if fused_score>1:
          fused_score=1
        fused_annotation = pred_1.copy()
        fused_annotation['score'] = fused_score
        fused_annotations.append(fused_annotation)

# Add any remaining predictions from Model 2
image_ids_1 = set([pred['image_id'] for pred in annotations_1])
for pred_2 in annotations_2:
    if pred_2['image_id'] not in image_ids_1:
        fused_annotation = pred_2.copy()
        fused_annotation['score'] = weight_2 * pred_2['score']
        fused_annotations.append(fused_annotation)



# Create COCO object for fused predictions
coco_dt = coco_gt.loadRes(fused_annotations)

```
